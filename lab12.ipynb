```python
import pandas as pd
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
df = pd.read_csv ('auto-mpg.csv')
x1 = df.weight
y = df.mpg
testX=np.arange(1500, 5500, 500, dtype=int)
slope, intercept, r,p,std_err =stats.linregress(x1,y)
Ymodel = intercept + (slope*x1)
Ymodel2 = intercept + (slope*testX)
df.plot.scatter(x='weight', y='mpg', label= 'Traning dataPoints', color='b')
plt.scatter(testX,Ymodel2,color='y',label='Traning dataPoints')
plt.plot(x1,Ymodel, color='r', label='OLS')
plt.legend()
plt.title((f'Linear Regression with OLS: Slope: {slope:.2f}, y-intercept: {intercept:.2f}, r: {r:.2f}'))
plt.text(3500,30,'Linear Regression line')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df = pd.read_csv ('housing.csv')
x=np.asarray(df.Longitude)
y=np.asarray(df.Latitude)
df=np.array(list(zip(x, y))).reshape(len(x), 2)
kmeans = KMeans(n_clusters=6).fit(df)
centers = np.array(kmeans.cluster_centers_)
print(kmeans.cluster_centers_)
centers = np.array(kmeans.cluster_centers_)
plt.title('k means Algo')
plt.scatter(x, y, c=kmeans.labels_.astype(float))    
plt.scatter(centers[:,0], centers[:,1], marker="*", color='r')
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.show()
```
